### 3. Benzersiz Müşteriler

```python
print("Hamamözü Benzersiz Müşteri Sayısı: ", df_tahakkuk["sozlesme_hesap_no"].nunique())
print("Gümüşhacıköy Benzersiz Müşteri Sayısı: ", df_tahakkuk_1["sozlesme_hesap_no"].nunique())
print("Göynücek Benzersiz Müşteri Sayısı: ", df_tahakkuk_2["sozlesme_hesap_no"].nunique())
```

### 4. Tüm Tahakkuk Verilerini Birleştirme

```python
df_tahakkuk_all=pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)

df_tahakkuk_all.shape
```

### 5. kwh Sütunu Eksik, Negatif, Aşırı Uç Değer Tespiti

```python
df_tahakkuk_all["kwh"].isna().sum()   # eksik değer bulunmamaktadır.
```

```python
(df_tahakkuk_all["kwh"]<0).sum()       # 151 adet negatif kayıt var. Fatura iptali ya da düzeltme olmuş olabilir.
```

```python
df_tahakkuk_all["kwh"].describe()
```

### 6. Hesap Sınıfına Göre Ortalama, Medyan ve Standart Sapma Tüketimleri

```python
hs_statistics= df_tahakkuk_all.groupby("Hesap Sınıfı")["kwh"].agg(ortalama_kwh="mean", median_kwh="median", std_kwh="std").sort_values("ortalama_kwh", ascending=False)
hs_statistics=hs_statistics.round(2)
hs_statistics.head(10)
```
